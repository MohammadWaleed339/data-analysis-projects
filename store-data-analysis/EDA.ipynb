```python
# Get the most profitable products
premium_product = pd.read_sql_query('''SELECT 
                                            pd.product_id,
                                            pd.supplier_id,
                                            r.customer_id,
                                            AVG(r.rating) AS avg_rating,            
                                            AVG(oi.price_at_purchase) AS avg_price_at_purchase,
                                            oi.order_item_id,
                                            SUM(oi.quantity) AS TotalQuantity,
                                            pd.product_name,
                                            pd.category,
                                            pd.price
                                        FROM order_items oi
                                        RIGHT JOIN products pd ON oi.product_id = pd.product_id
                                        RIGHT JOIN reviews r ON pd.product_id = r.product_id
                                        WHERE oi.price_at_purchase > 0
                                        GROUP BY r.customer_id, pd.product_id, pd.product_name, pd.category, pd.price, pd.supplier_id,oi.order_item_id
                                        ORDER BY avg_price_at_purchase DESC;

''', conn)
```

```python
# get details of customers who is buying a lot
premium_customer = pd.read_sql_query(''' SELECT 
                                             c.customer_id,
                                             c.first_name,
                                             c.last_name,
                                             c.email,
                                             SUM(o.total_price) as SumOfTotalPrice,
                                             SUM(p.amount) as TotalPayment,
                                             AVG(r.rating) as AverageRating
                                         FROM customers c
                                         LEFT JOIN orders o
                                         ON c.customer_id = o.customer_id
                                         LEFT JOIN payment p
                                         ON o.order_id = p.order_id
                                         LEFT JOIN reviews r
                                         ON r.customer_id = c.customer_id
                                         GROUP BY c.customer_id, c.first_name, c.last_name, c.email
                                         ORDER BY SumOfTotalPrice  desc
                                ''', conn)
```

```python
final_table = pd.read_sql_query(''' WITH customer_table AS (SELECT 
                                                                 c.customer_id,
                                                                 c.first_name,
                                                                 c.last_name,
                                                                 c.email,
                                                                 SUM(o.total_price) as LifeTimeCustomerValue,
                                                                 SUM(p.amount) as LifeTimePayment,
                                                                 AVG(r.rating) as AverageRating
                                                             FROM customers c
                                                             LEFT JOIN orders o
                                                             ON c.customer_id = o.customer_id
                                                             LEFT JOIN payment p
                                                             ON o.order_id = p.order_id
                                                             LEFT JOIN reviews r
                                                             ON r.customer_id = c.customer_id
                                                             GROUP BY c.customer_id, c.first_name, c.last_name, c.email
                                                             ORDER BY LifeTimeCustomerValue  desc),
                                            
                                            order_table AS (SELECT 
                                                                pd.product_id,
                                                                pd.supplier_id,
                                                                r.customer_id,
                                                                AVG(r.rating) AS avg_rating,            
                                                                AVG(oi.price_at_purchase) AS AvgItemPriceAtPurchase,
                                                                oi.order_item_id,
                                                                SUM(oi.quantity) AS TotalQuantity,
                                                                pd.product_name,
                                                                pd.category,
                                                                pd.price
                                                            FROM order_items oi
                                                            RIGHT JOIN products pd ON oi.product_id = pd.product_id
                                                            RIGHT JOIN reviews r ON pd.product_id = r.product_id
                                                            WHERE oi.price_at_purchase > 0
                                                            GROUP BY r.customer_id, pd.product_id, pd.product_name, pd.category, pd.price, pd.supplier_id,oi.order_item_id
                                                            ORDER BY AvgItemPriceAtPurchase DESC),
                                          shipment_table AS (SELECT 
                                                                s.shipment_id,
                                                                o.order_date,
                                                                o.customer_id,
                                                                o.total_price,
                                                                s.shipment_date,
                                                                s.carrier,
                                                                s.delivery_date,
                                                                s.shipment_status,
                                                                oi.product_id
                                                            FROM shipments s
                                                            JOIN orders o
                                                            ON s.order_id = o.order_id
                                                            JOIN order_items oi
                                                            ON oi.order_id = o.order_id
                                                            )
                                        SELECT 
                                            ot.product_id,
                                            ot.supplier_id,
                                            ct.customer_id,
                                            ct.LifeTimeCustomerValue,
                                            ct.LifeTimePayment,
                                            st.carrier,
                                            ot.avg_rating,
                                            ot.AvgItemPriceAtPurchase,
                                            ot.TotalQuantity,
                                            ot.product_name,
                                            st.delivery_date,
                                            shipment_date,
                                            ot.category,
                                            st.shipment_status,
                                            ot.price AS ShippedProductPrice
                                        FROM customer_table ct
                                        RIGHT JOIN order_table ot
                                        ON ot.customer_id = ct.customer_id
                                        JOIN shipment_table st
                                        ON st.customer_id = ot.customer_id
                                        
                                        
                                                            ''',conn)
final_table                                           
                                        
```

```python
table_query = ''' CREATE TABLE final_table (customer_id INT, 
                                            product_id INT,
                                            supplier_id INT,
                                            LifeTimeCustomerValue FLOAT,
                                            LifeTimePayment FLOAT,
                                            avg_rating INT,
                                            AvgItemPriceAtPurchase FLOAT,
                                            carrier VARCHAR(10),
                                            TotalQuantity INT,
                                            product_name VARCHAR(20),
                                            delivery_date DATE,
                                            shipment_date DATE,
                                            category VARCHAR(20),
                                            shipment_status VARCHAR(20),
                                            ProductPrice FLOAT,
                                            PRIMARY KEY(customer_id,product_id))
                                            '''
```
